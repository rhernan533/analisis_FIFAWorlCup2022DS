# 2- Anlaisis de datos

```python
df2['Posesión E1'] = df2['Posesión E1'].map(lambda x: x.replace('%','')).astype(float)
df2['Posesión E1']
```

```python
df2['Posesión nula'] = df2['Posesión nula'].map(lambda x: x.replace('%','')).astype(float)
df2['Posesión nula']
```

```python
equipo1 = df2.groupby('Equipo 1').agg(
    {'Goles E1':"sum",
     'Intentos de remate E1' : "sum",
     'Posesión E1' : "mean",
     'Asistencias E1' : "sum",
     'Faltas E1' : "sum",
     'Tarjetas amarillas E1' : "sum",
     'Tarjetas rojas E1' : "sum",
     'Posesión nula' : "mean"}).reset_index()
equipo1.head()
```

```python
df2['Posesión E2'] = df2['Posesión E2'].map(lambda x: x.replace('%','')).astype(float)
df2['Posesión E2']
```

```python
equipo2 = df2.groupby('Equipo 2').agg(
    {'Goles E2':"sum",
     'Intentos de remate E2' : "sum",
     'Posesión E2' : "mean",
     'Asistencias E2' : "sum",
     'Faltas E2' : "sum",
     'Tarjetas amarillas E2' : "sum",
     'Tarjetas rojas E2' : "sum",
     'Posesión nula' : "mean"}).reset_index()
equipo2.head()
```

```python
estadist_equip = pd.DataFrame(columns= ['seleccion', 'total_goles', 'total_intentos', 'posesion_promedio', 'total_asistencias',
                             'total_faltas', 'total_amarillas', 'total_rojas', 'promedio_posesion_nula'])

estadist_equip

```

```python
estadist_equip
```

```python
estadist_equip['total_intentos'] = equipo1['Intentos de remate E1'] + equipo2['Intentos de remate E2']
estadist_equip['posesion_promedio'] = (equipo1['Posesión E1'] + equipo2['Posesión E2'])/2
estadist_equip['total_asistencias'] = equipo1['Asistencias E1'] + equipo2['Asistencias E2']
estadist_equip['total_faltas'] = equipo1['Faltas E1'] + equipo2['Faltas E2']
estadist_equip['total_amarillas'] = equipo1['Tarjetas amarillas E1'] + equipo2['Tarjetas amarillas E2']
estadist_equip['total_rojas'] = equipo1['Tarjetas rojas E1'] + equipo2['Tarjetas rojas E2']
estadist_equip['promedio_posesion_nula'] = (equipo1['Posesión nula'] + equipo2['Posesión nula'])/2

estadist_equip
```

## b- Graficos

```python
top_7_goles = estadist_equip.sort_values('total_goles',ascending=False).head(7).set_index('seleccion', drop= False)
top_7_goles
```

### Goles totales, intentos de goles y efectividad

Haremos una grafica de los 7 equipos que marcaron mas goles.
Analizaremos la cantidad de llegadas al arco.
Y su efectividad en goles respecto a la cantidad de llegadas

```python
selecciones = top_7_goles['seleccion'].tolist()
goles = top_7_goles['total_goles'].tolist()
intentos_gol = top_7_goles['total_intentos'].tolist()

x = np.arange(len(selecciones))  # the label locations
ancho = 0.35  # the width of the bars

fig, ax = plt.subplots()

graf1 = ax.bar(x - ancho/2, goles, ancho, label = 'Goles totales')
graf2 = ax.bar(x + ancho/2, intentos_gol, ancho, label = 'Intentos de gol totales')


ax.set_title('Goles TOP 7 selecciones')
ax.set_xlabel('Selecciones')
ax.set_xticks(x, selecciones, fontsize=7)
ax.set_ylim(0, 120)
ax.legend()


def autolabel(barras):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for barra in barras:
        height = barra.get_height()
        ax.annotate('{}'.format(height),
                    xy=(barra.get_x() + barra.get_width() / 2, height-7),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        
for i,j, posicion in zip(goles, intentos_gol, graf1):
    efectividad = round((i/j)*100, 2)
    x = posicion.get_x()
    ax.annotate('{}%'.format(efectividad), xy= (x, i+2), fontsize=6)
    

        
autolabel(graf1)
autolabel(graf2)
        


plt.show
```

### Relación entre faltas cometidas y tarjetas

```python
faltas1 = df2.groupby('Equipo 1', as_index=False).agg({"Tarjetas amarillas E1" : "mean", "Tarjetas rojas E1" : "mean",
                                   "Faltas E1" : "mean"}).rename(columns={"Tarjetas amarillas E1":"amarillas1",
                                                                                    "Tarjetas rojas E1":"rojas1",
                                                                                    "Faltas E1":"faltas_cometidas1",
                                                                                    "Equipo 1" : "seleccion"}) .round(2)
faltas1.head()
```

```python
faltas2 = df2.groupby('Equipo 2', as_index=False).agg({"Tarjetas amarillas E2" : "mean", "Tarjetas rojas E2" : "mean",
                                   "Faltas E2" : "mean"}).rename(columns={"Tarjetas amarillas E2":"amarillas2",
                                                                                    "Tarjetas rojas E2":"rojas2",
                                                                                    "Faltas E2":"faltas_cometidas2",
                                                                                    "Equipo 2" : "seleccion"}) .round(2)
faltas2.head()
```

```python
fouls_and_cards = faltas1.merge(faltas2, on = 'seleccion')
fouls_and_cards.head()
```

```python
fouls_and_cards = fouls_and_cards.assign(amarillas_por_partido = (fouls_and_cards.amarillas1 + fouls_and_cards.amarillas2) / 2 )
fouls_and_cards = fouls_and_cards.assign(rojas_por_partido = (fouls_and_cards.rojas1 + fouls_and_cards.rojas2) / 2 )
fouls_and_cards = fouls_and_cards.assign(faltas_por_partido = (fouls_and_cards.faltas_cometidas1 + fouls_and_cards.faltas_cometidas2) / 2 )
fouls_and_cards = fouls_and_cards[['seleccion', 'faltas_por_partido', 'amarillas_por_partido' , 'rojas_por_partido']].round(2)
```

```python
fouls_and_cards
```

```python
fouls_and_cards.sort_values('faltas_por_partido',ascending = False).head().set_index('seleccion')
```

```python
ax2 = sns.scatterplot(data = fouls_and_cards, x = 'faltas_por_partido', y = 'amarillas_por_partido', hue='seleccion', 
size="faltas_por_partido",sizes=(50, 500)).set_title("Fouls and cards\n")
plt.legend(bbox_to_anchor=(1.14, 0.5), loc='right', borderaxespad=0)

sns.despine()
```

```python
fouls_and_cards[['seleccion', 'faltas_por_partido', 'amarillas_por_partido']].set_index('seleccion').plot(kind='bar', title = 'Fouls and Yellow cards per game\n')

plt.show()
```

Arabia Saudí y Serbia son líderes en número de tarjetas amarillas por partido.
